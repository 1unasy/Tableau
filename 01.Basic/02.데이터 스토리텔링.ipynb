# < 데이터 스토리텔링 >
### (1) 이해관계자 & 의사결정자와 커뮤니케이션하기

- 데이터 스토리텔링 프레임워크를 활용해 **그들의 언어로 커뮤니케이션하기**


### @ 스토리의 구성 요소
#### 단순히 나열된 정보가 아니라 흐름이 있는 이야기로 만들어서 뇌 활성화시키기
    - 상황(Context) → 문제(Problem) → 해결(Action)

### @ 스토리텔링 체크리스트
#### - 청자 : 이해관계자들과 협업 / 의사결정권자에게 발표
#### - 주제 : 하나의 핵심 문장과 명확한 액션 아이템
#### - 근거 : 데이터 시각화를 통해 그 근거를 직관적으로 전달

### @ 유형별 커뮤니케이션 방법 (분석가의 입장)
#### - 현업과 문제를 찾아나가는 법
    - 1:1 인터뷰 하며 진짜 문제 찾기
        - 암묵지를 이끌어내기 위해 why?를 계속해서 반복한다.
        - Yes or No 로 답변할 수 없는 질문이 더 좋은 대답을 이끌어낸다.
        - 인터뷰에서 나온 내용이 아니더라도 관찰 후 Pain Point를 찾아보고 이게 맞는지 질문한다.             
#### - 현업의 협조를 이끌어내는 전략
    - 신뢰와 친밀감 쌓기
        - 미리 분석한 데이터 시각화 자료들을 조직에 꾸준히 공유한다.
        - 회사에서 편안한 1:1 자리를 많이 만든다.                
#### - 의사결정자(C-level)와 소통하기
    - 핵심 포인트만 빠르게 전달하기
        - 결론 / 이유 & 예시 / 추천하는 액션
    - 의사결정자들이 성과를 평가받는 방법 고려하기
        - 제안은 6가지 핵심 성과 지표 중 한 영역을 개선해야 한다.
            - Money
                - 매출(Revenue) ▲ / 비용(Costs) ▼
            - Market
                - 시장점유율 (Market Share) ▲ / 제품이 시장에 도달하는 데 걸리는 시간 (Time to Market) ▼
            - Exposure
                - 유지 (Retention) ▲ / 리스크 (Risk) ▼
    - 의사결정자가 선호하는 커뮤니케이션 방식 파악하기
        - 대면 or 서면(일대일 면담, PT, 제안서, 이메일)
        - 가치관(평소에 가지는 의견, 관심 영역)
        - 제시가능한 반론 대처방안

### (2) 핵심 메시지 전달하기
### @ 스토리의 3단 구성
#### - 컨텍스트 설명 (현재 상황)
#### - 핵심 메시지 (문제와 기회)
#### - 추천하는 액션
### @ 핵심 메시지 전달하기
- **point** : 핵심 메시지 / 액션 아이템 **+ 득과 실** : 예상 변화 수치 / 시간과 비용

### @ 제안서의 구조
- 데이터 스토리 : 핵심 요약 메시지
    - 하위 액션 아이템
    - Action 1 / Action 2 / Action 3
        - What : 무엇을 해야 하는가?
        - Why : 왜 그것이 변해야 하는가?
        - How : 어떻게 긍정적인 변화를 만들 것인가?
