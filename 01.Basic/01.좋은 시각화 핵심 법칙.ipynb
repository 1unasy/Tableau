# < 좋은 시각화 핵심 법칙 >

## 01. 시그널은 최대화 노이즈는 최소화
- 사람이 한 번에 처리할 수 있는 정보량에 한계가 있기 때문에 '정보 전달의 효율성'을 생각해야 함
- 표현하고자 하는 것에 대한 시그널 최대화 이외의 것들 최소화
    
    ▷ 데이터 잉크 비율을 높이자
    - 표현하고자하는 데이터 외에는 불필요한 장식을 없애 심플하게 만들자 for 커뮤니케이션

## 02. 시각 속성 활용하기 (길이, 너비, 사이즈, 채도 등) 
- 인간의 눈이 미묘한 차이에도 정확하게 비교할 수 있는 순서
    - **위치 > 길이 > 방향 > 각도 > 면적 > 부피 > 채도**
	
    
    ▷ 막대 차트(길이)는 미묘한 길이도 판별할 수 있어 가장 일반적으로 사용
    ▷ 파이 차트(각도)와 영역 차트(면적)는 인간의 눈에 정확하게 인식되기 어려움
        - 파이 차트를 사용하고 싶다면, '도넛 차트' 활용하기

## 03. 데이터 시각화의 목적
### @ 커뮤니케이션(communication)
##### Explanatory Analysis (설명형 분석) 대상
- 질문에 대한 **핵심 메시지**가 명확한가?
- **Action Item** (행동을 촉진하는 내용) 이 있는가?
- 상대방과 나 사이에 공유된 **컨텍스트**를 고려했는가?
- 인사이트를 얼마나 **이해하기 쉬운 형태**로 전달하는가?

## 04. 실전 데이터 시각화 테크닉
#### 1. 3D 차트는 쓰지 않는다!
    - 정확한 사실이 전달되지 않을 뿐만 아니라 왜곡된 정보가 전달될 가능성이 크다.
#### 2. 막대그래프의 축은 반드시 0부터 시작한다!
    - 그렇지 않으면 차이가 과도하게 강조되어 사실을 왜곡하게 된다.
#### 3. 꺾은선 그래프의 축은 잘라내도 괜찮다!
    - 트렌드를 나타내는 꺾은선 그래프의 가장 중요한 역할은 선의 기울기로 경향을 파악하는 것이기 때문이다.
#### 4. 색은 강조하고 싶은 요소에만 사용한다!
    - 색상에 시각화의 목적을 반영해서 보는 사람에게 명확하게 메시지를 전달할 수 있도록 한다.
#### 5. 이중축은 혼란의 여지가 있다!
    - 어느 쪽 축이 막내이고 어느쪽 축이 선인지 불분명해 이해하기 어렵다.
#### 6. 산포도는 가로축에 원인, 세로축에 결과를 표시한다.
    - 산포도의 상관관계가 인관관계를 나타내지는 않지만, 일반론적으로 성립된다고 생각하는 경우 해석하기 쉽게 하기 위해서 **선행지표 : x축, 후행지표 : y축**을 기본으로 한다.
#### 7. 시간 축은 가로축이 기본이다!
    - 시간을 가로로 배치하는 편이 경향을 인식하기 훨씬 편리하다.
